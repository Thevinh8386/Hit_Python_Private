Câu 5:
1. Cách khai báo biến trong Python:
Python không cần khai báo kiểu dữ liệu trước, chỉ cần gán giá trị cho biến.
Ví dụ:
x = 19          
name = "Vinh"   
is_student = True  

2. Các toán tử trong Python:
Toán tử số học:
      +, -, *, /, // (chia lấy nguyên), % (chia lấy dư), ** (lũy thừa)
Toán tử so sánh:
      ==, !=, >, <, >=, <=
Toán tử logic:
      and, or, not
Toán tử gán:
      =, +=, -=, *=, /=, ...
Toán tử bit:
      &, |, ^, ~, <<, >>

3. Các kiểu dữ liệu trong Python:
Số nguyên (int):
      Ví dụ: x = 5, y = 7, ...
Số thực (float):
      Ví dụ: y = 3.14, x = 4.25, ...
Chuỗi (str):
      Ví dụ: name = "Vinh"
Boolean (bool): True hoặc False
Danh sách (list):
      Ví dụ: my_list = [1, 2, 3]
Tuple:
      Ví dụ: my_tuple = (1, 2, 3)
Từ điển (dict):
      Ví dụ: my_dict = {"name": "Vinh", "age": 19}
Set:
      Ví dụ: my_set = {1, 2, 3}

```python
#Ví dụ về khai báo biến
age = 19         # Biến nguyên
name = "Vinh"   # Biến chuỗi
is_student = True  # Biến boolean
```

```python
#Toán tử số học
a = 10
b = 3
print(a + b)   # 13
print(a - b)   # 7
print(a * b)   # 30
print(a / b)   # 3.333...
print(a // b)  # 3 (chia lấy nguyên)
print(a % b)   # 1 (chia lấy dư)
print(a ** b)  # 1000 (lũy thừa 10^3)
```

```python
#Toán tử so sánh
x = 7
y = 10
print(x == y)   # False
print(x != y)   # True
print(x > y)    # False
print(x < y)    # True
print(x >= y)   # False
print(x <= y)   # True
```

```python
#Toán tử logic
x = True
y = False
print(x and y)   # False
print(x or y)    # True
print(not x)     # False
```

```python
#Toán tử gán
x = 5
x += 3    # x = x + 3 -> 8
x -= 2    # x = x - 2 -> 6
x *= 4    # x = x * 4 -> 24
x /= 6    # x = x / 6 -> 4.0
print("x = ", x)
```

```python
#Toán tử bit
a = 60    # 60 = 0011 1100
b = 13    # 13 = 0000 1101
print(a & b)   # 12 (0000 1100)
print(a | b)   # 61 (0011 1101)
print(a ^ b)   # 49 (0011 0001)
print(~a)      # -61 (bit đảo ngược của a)
print(a << 2)  # 240 (dịch trái 2 bit, 0011 1100 -> 1111 0000)
print(a >> 2)  # 15 (dịch phải 2 bit, 0011 1100 -> 0000 1111)
```

```python
#Ví dụ về các kiểu dữ liệu

# Số nguyên (int)
age = 19
print("age:", age, "| type:", type(age))

# Số thực (float)
height = 5.7
print("height:", height, "| type:", type(height))

# Chuỗi (str)
name = "Vinh"
print("name:", name, "| type:", type(name))

# Boolean (bool)
is_student = True
print("is_student:", is_student, "| type:", type(is_student))

# Danh sách (list)
fruits = ["apple", "banana", "mango"]
print("fruits:", fruits, "| type:", type(fruits))

# Tuple
my_tuple = (10, 20)
print("my_tuple:", my_tuple, "| type:", type(my_tuple))

# Từ điển (dict)
person = {"name": "Vinh", "age": 19}
print("person:", person, "| type:", type(person))

# Set
my_numbers = {1, 2, 3, 4, 5}
print("my_numbers:", my_numbers, "| type:", type(my_numbers))
```
